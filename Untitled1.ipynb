```python
import json
import statistics
import warnings
import numpy as np
import pandas as pd
import scipy.sparse as sparse
from scipy.sparse.linalg import spsolve
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
warnings.filterwarnings('ignore')
```

```python
print("\n" + "═"*65)
print("  STEP 1 : Loading & Cleaning Trek Data")
print("═"*65)

with open('final_trek_dataset_with_images.json') as f:
    raw = json.load(f)
treks = raw['treks']

# Fix duplicate trek_id
seen = set()
for t in treks:
    if t['trek_id'] in seen and t['name'] == 'Putha Hiunchuli Trek (Expedition)':
        t['trek_id'] = 'puh001'
    seen.add(t['trek_id'])

# Impute missing base_camp_altitude_m with median
alts = [t['base_camp_altitude_m'] for t in treks if t.get('base_camp_altitude_m')]
med_alt = statistics.median(alts)
for t in treks:
    if not t.get('base_camp_altitude_m'):
        t['base_camp_altitude_m'] = med_alt

print(f"  ✓ {len(treks)} treks loaded")
print(f"  ✓ Duplicate trek ID fixed")
print(f"  ✓ Missing altitude values imputed with median = {med_alt}m")


# ════════════════════════════════════════════════════════════════════════════
#  STEP 2 — ENCODE TREK FEATURES INTO NUMBERS
# ════════════════════════════════════════════════════════════════════════════
#
#  WHY? The model needs all features as numbers.
#  We use ordinal encoding — categories that have an order get
#  a number that reflects their rank (easy=1, moderate=2 … very challenging=6)
#
print("\n" + "═"*65)
print("  STEP 2 : Encoding Trek Features")
print("═"*65)

difficulty_map    = {'very easy':1,'easy':2,'moderate':3,
                     'moderate to strenuous':4,'challenging':5,'very challenging':6}
fitness_map       = {'basic':1,'moderate':2,'good':3,'excellent':4}
avail_map         = {'very limited':1,'limited':2,'moderate':3,'good':4,'excellent':5}
network_map       = {'none':1,'very limited':2,'limited':3,
                     'moderate':4,'good':5,'excellent':6}
risk_map          = {'very low':1,'low':2,'moderate':3,'high':4,'very high':5}
altitude_risk_map = {'none':0,'low':1,'moderate':2,'high':3,'very high':4}

trek_ids   = [t['trek_id'] for t in treks]
trek_names = {t['trek_id']: t['name'] for t in treks}

# Build the trek feature matrix — each trek becomes a row of numbers
trek_features = []
for t in treks:
    trek_features.append([
        difficulty_map[t['difficulty']],
        fitness_map[t['fitness_level_required']],
        risk_map[t['risk_level']],
        altitude_risk_map[t['altitude_sickness_risk']],
        avail_map[t['water_availability']],
        avail_map[t['food_availability']],
        network_map[t['mobile_network']],
        int(t['permits_required']),
        int(t['guide_mandatory']),
        int(t['evacuation_possible']),
        t['duration_days'],
        t['distance_km'],
        t['max_altitude_m'],
        t['altitude_gain_m'],
        t['daily_trek_hours'],
        t['estimated_cost_min_usd'],
        t['estimated_cost_max_usd'],
        t['temperature_min'],
        t['temperature_max'],
        t['nearest_medical_facility_km'],
        t['average_rating'],
        t['popularity_score'],
        t['group_size_min'],
        t['group_size_max'],
    ])

trek_feature_df = pd.DataFrame(trek_features, index=trek_ids, columns=[
    'difficulty','fitness','risk','altitude_risk',
    'water','food','network',
    'permits','guide','evacuation',
    'duration_days','distance_km','max_altitude_m','altitude_gain_m',
    'daily_hours','cost_min','cost_max',
    'temp_min','temp_max','nearest_medical_km',
    'avg_rating','popularity','group_min','group_max'
])

# Scale all features to 0–1 so no single feature dominates
scaler = MinMaxScaler()
trek_feature_scaled = scaler.fit_transform(trek_feature_df)

print(f"  ✓ {trek_feature_df.shape[1]} features encoded per trek")
print(f"  ✓ All features scaled to [0, 1]")


# ════════════════════════════════════════════════════════════════════════════
#  STEP 3 — GENERATE SYNTHETIC USER-TREK INTERACTION MATRIX
# ════════════════════════════════════════════════════════════════════════════
#
#  WHY SYNTHETIC USERS?
#  Real recommendation systems need user history (who rated what).
#  Since we don't have real users yet, we CREATE realistic fake users
#  based on different trekker profiles. This simulates the interaction
#  matrix that ALS needs to train on.
#
#  Each synthetic user has:
#    - A preferred difficulty, budget, duration, fitness level, region
#    - Ratings are computed as a score from 0–5 based on how well
#      each trek matches their preferences
#
print("\n" + "═"*65)
print("  STEP 3 : Generating Synthetic User-Trek Interaction Matrix")
print("═"*65)

np.random.seed(42)

# Define realistic trekker profiles
USER_PROFILES = [
    # (profile_name, difficulty_pref, budget_max, duration_max, fitness, region_pref)
    ('Beginner Budget',        1, 300,  7,  'basic',    None),
    ('Casual Explorer',        2, 500,  10, 'moderate', None),
    ('Weekend Trekker',        2, 400,  5,  'moderate', 'Kathmandu Valley'),
    ('Nature Lover',           3, 700,  12, 'good',     'Annapurna'),
    ('Photography Enthusiast', 3, 800,  14, 'good',     None),
    ('Cultural Trekker',       2, 600,  10, 'moderate', 'Mustang'),
    ('Family Trekker',         2, 1000, 10, 'moderate', None),
    ('Fitness Enthusiast',     4, 1000, 14, 'good',     'Manaslu'),
    ('Adventure Seeker',       5, 2000, 18, 'excellent','Khumbu'),
    ('High Altitude Lover',    5, 2500, 21, 'excellent','Khumbu'),
    ('Budget Backpacker',      3, 400,  14, 'good',     None),
    ('Luxury Trekker',         3, 5000, 14, 'good',     'Annapurna'),
    ('Solo Trekker',           4, 1500, 16, 'excellent',None),
    ('Remote Explorer',        5, 3000, 20, 'excellent','Dolpo'),
    ('Himalayan Enthusiast',   6, 4000, 25, 'excellent','Dhaulagiri'),
]

# Generate 500 users by sampling from profiles with noise
N_USERS = 500
users = []
for i in range(N_USERS):
    base = USER_PROFILES[i % len(USER_PROFILES)]
    noise_diff   = np.random.randint(-1, 2)
    noise_budget = np.random.uniform(0.7, 1.4)
    noise_dur    = np.random.randint(-2, 3)
    users.append({
        'user_id':    f'user_{i:04d}',
        'profile':    base[0],
        'diff_pref':  max(1, min(6, base[1] + noise_diff)),
        'budget_max': int(base[2] * noise_budget),
        'dur_max':    max(1, base[3] + noise_dur),
        'fitness':    fitness_map[base[4]],
        'region':     base[5],
    })

user_ids = [u['user_id'] for u in users]

# Build interaction matrix: rows=users, cols=treks
# Rating = how well trek matches user preferences (0–5 scale)
def compute_rating(user, trek):
    score = 0.0
    weights = 0.0

    # Difficulty match (most important — weight 3)
    diff_diff = abs(user['diff_pref'] - difficulty_map[trek['difficulty']])
    score   += max(0, 5 - diff_diff * 1.5) * 3
    weights += 3 * 5

    # Budget match (weight 2.5)
    cost_mid = (trek['estimated_cost_min_usd'] + trek['estimated_cost_max_usd']) / 2
    if cost_mid <= user['budget_max']:
        score   += 5 * 2.5
    else:
        over = (cost_mid - user['budget_max']) / max(user['budget_max'], 1)
        score   += max(0, 5 - over * 3) * 2.5
    weights += 2.5 * 5

    # Duration match (weight 2)
    if trek['duration_days'] <= user['dur_max']:
        score   += 5 * 2
    else:
        over = trek['duration_days'] - user['dur_max']
        score   += max(0, 5 - over * 0.5) * 2
    weights += 2 * 5

    # Fitness match (weight 2)
    fit_diff = abs(user['fitness'] - fitness_map[trek['fitness_level_required']])
    score   += max(0, 5 - fit_diff * 1.2) * 2
    weights += 2 * 5

    # Region preference (weight 1.5)
    if user['region'] and user['region'] == trek['region']:
        score   += 5 * 1.5
    elif user['region'] is None:
        score   += 3 * 1.5   # neutral
    else:
        score   += 1 * 1.5
    weights += 1.5 * 5

    # Trek quality (weight 1)
    quality = (trek['average_rating'] - 4.2) / 0.8 * 5  # scale 4.2–5.0 → 0–5
    score   += quality * 1
    weights += 1 * 5

    # Normalise to 0–5
    rating = round((score / weights) * 5, 2)
    # Add slight noise to simulate real-world variability
    rating += np.random.normal(0, 0.15)
    return float(np.clip(rating, 0.5, 5.0))

# Only generate interactions for ~40% of user-trek pairs (sparse matrix)
# Real systems are very sparse — most users haven't rated most items
SPARSITY = 0.40
interaction_data = []
for user in users:
    for t in treks:
        if np.random.random() < SPARSITY:
            rating = compute_rating(user, t)
            interaction_data.append((user['user_id'], t['trek_id'], rating))

interaction_df = pd.DataFrame(interaction_data, columns=['user_id','trek_id','rating'])

# Create user/trek index maps for the matrix
user_idx  = {uid: i for i, uid in enumerate(user_ids)}
trek_idx  = {tid: i for i, tid in enumerate(trek_ids)}
n_users   = len(user_ids)
n_treks   = len(trek_ids)

# Build sparse matrix (memory-efficient for large datasets)
rows = interaction_df['user_id'].map(user_idx).values
cols = interaction_df['trek_id'].map(trek_idx).values
vals = interaction_df['rating'].values
interaction_matrix = sparse.csr_matrix((vals, (rows, cols)),
                                        shape=(n_users, n_treks))

density = interaction_df.shape[0] / (n_users * n_treks) * 100
print(f"  ✓ {N_USERS} synthetic users generated from {len(USER_PROFILES)} profiles")
print(f"  ✓ {len(interaction_data):,} user-trek interactions created")
print(f"  ✓ Matrix shape : {n_users} users × {n_treks} treks")
print(f"  ✓ Matrix density : {density:.1f}% (rest are unknowns ALS will predict)")


# ════════════════════════════════════════════════════════════════════════════
#  STEP 4 — TRAIN / TEST SPLIT
# ════════════════════════════════════════════════════════════════════════════
#
#  WHY SPLIT?
#  We hide 20% of the known ratings from the model during training.
#  After training, we ask the model to PREDICT those hidden ratings
#  and compare to the real ones — this tells us how accurate the model is.
#
print("\n" + "═"*65)
print("  STEP 4 : Train / Test Split  (80% train, 20% test)")
print("═"*65)

train_df, test_df = train_test_split(interaction_df, test_size=0.2, random_state=42)

# Build train matrix
rows_tr = train_df['user_id'].map(user_idx).values
cols_tr = train_df['trek_id'].map(trek_idx).values
vals_tr = train_df['rating'].values
train_matrix = sparse.csr_matrix((vals_tr, (rows_tr, cols_tr)),
                                   shape=(n_users, n_treks))

print(f"  ✓ Training set   : {len(train_df):,} interactions")
print(f"  ✓ Test set       : {len(test_df):,} interactions")
```

```python
class ALSRecommender:
    """
    Alternating Least Squares Matrix Factorization Recommender.

    Parameters:
      n_factors   : number of latent factors  (hidden dimensions)
      n_iterations: how many ALS training rounds
      lambda_reg  : regularisation strength (prevents overfitting)
      verbose     : print training progress
    """

    def __init__(self, n_factors=20, n_iterations=20, lambda_reg=0.1, verbose=True):
        self.n_factors    = n_factors
        self.n_iterations = n_iterations
        self.lambda_reg   = lambda_reg
        self.verbose      = verbose
        self.U = None   # User  latent matrix
        self.V = None   # Trek  latent matrix
        self.train_losses = []

    def fit(self, R):
        """
        Train the ALS model on interaction matrix R.
        R : scipy sparse matrix  (n_users × n_treks)
        """
        n_users, n_treks = R.shape
        k = self.n_factors

        # ── Initialise U and V with small random values ──────────────────
        # Random start gives the model something to improve from
        self.U = np.random.normal(scale=0.1, size=(n_users, k))
        self.V = np.random.normal(scale=0.1, size=(n_treks, k))

        R_csr = R.tocsr()   # user  × trek  (for updating U)
        R_csc = R.tocsc()   # trek  × user  (for updating V)
        I_k   = np.eye(k)   # identity matrix for regularisation

        if self.verbose:
            print(f"\n  Training ALS — {n_users} users, {n_treks} treks, "
                  f"{k} latent factors, {self.n_iterations} iterations")
            print(f"  {'Iter':>5}  {'Train RMSE':>12}")
            print(f"  {'─'*5}  {'─'*12}")

        for iteration in range(self.n_iterations):

            # ── Step A: Fix V, solve for each user row in U ───────────────
            # For user i:  U[i] = (V^T V + λI)^-1  V^T r_i
            # where r_i = ratings given by user i (only observed entries)
            VtV = self.V.T @ self.V
            for i in range(n_users):
                # Get the treks this user has rated
                start, end = R_csr.indptr[i], R_csr.indptr[i+1]
                rated_treks = R_csr.indices[start:end]
                rated_vals  = R_csr.data[start:end]
                if len(rated_treks) == 0:
                    continue
                V_i   = self.V[rated_treks]
                A     = V_i.T @ V_i + self.lambda_reg * I_k
                b     = V_i.T @ rated_vals
                self.U[i] = np.linalg.solve(A, b)

            # ── Step B: Fix U, solve for each trek row in V ───────────────
            # For trek j:  V[j] = (U^T U + λI)^-1  U^T r_j
            UtU = self.U.T @ self.U
            for j in range(n_treks):
                start, end = R_csc.indptr[j], R_csc.indptr[j+1]
                rated_users = R_csc.indices[start:end]
                rated_vals  = R_csc.data[start:end]
                if len(rated_users) == 0:
                    continue
                U_j   = self.U[rated_users]
                A     = U_j.T @ U_j + self.lambda_reg * I_k
                b     = U_j.T @ rated_vals
                self.V[j] = np.linalg.solve(A, b)

            # ── Compute training RMSE ─────────────────────────────────────
            preds = np.array([
                self.U[user_idx[row['user_id']]] @ self.V[trek_idx[row['trek_id']]]
                for _, row in train_df.iterrows()
            ])
            preds   = np.clip(preds, 0.5, 5.0)
            rmse    = np.sqrt(mean_squared_error(train_df['rating'], preds))
            self.train_losses.append(rmse)

            if self.verbose and (iteration % 5 == 0 or iteration == self.n_iterations - 1):
                print(f"  {iteration+1:>5}  {rmse:>12.4f}")

        return self

    def predict(self, user_i, trek_j):
        """Predict rating for user index user_i and trek index trek_j."""
        return float(np.clip(self.U[user_i] @ self.V[trek_j], 0.5, 5.0))

    def recommend(self, user_id, top_n=10, exclude_seen=True):
        """
        Return top_n trek recommendations for a given user_id.
        exclude_seen : don't recommend treks the user has already rated
        """
        ui = user_idx[user_id]
        scores = self.U[ui] @ self.V.T   # dot product with all treks at once
        scores = np.clip(scores, 0.5, 5.0)

        if exclude_seen:
            seen_treks = set(
                interaction_df[interaction_df['user_id'] == user_id]['trek_id']
            )
            for tid, ji in trek_idx.items():
                if tid in seen_treks:
                    scores[ji] = -1   # push seen treks to the bottom

        top_indices = np.argsort(scores)[::-1][:top_n]
        return [(trek_ids[ji], trek_names[trek_ids[ji]], round(scores[ji], 3))
                for ji in top_indices]

    def recommend_for_new_user(self, preferences, top_n=10):
        """
        Recommend treks for a BRAND NEW user (no history).
        We create a user vector by averaging the vectors of similar users
        based on their preference profile.

        preferences dict keys:
          difficulty  : 1–6
          budget_max  : int  (USD)
          duration_max: int  (days)
          fitness     : 1–4
        """
        # Find the 10 users most similar in preference to the new user
        pref_scores = []
        for u in users:
            sim  = 0
            sim -= abs(u['diff_pref']  - preferences.get('difficulty', 3))   * 2
            sim -= abs(u['budget_max'] - preferences.get('budget_max', 1000)) / 500
            sim -= abs(u['dur_max']    - preferences.get('duration_max', 10)) * 0.5
            sim -= abs(u['fitness']    - preferences.get('fitness', 2))       * 1.5
            pref_scores.append(sim)

        top_similar_idx = np.argsort(pref_scores)[::-1][:10]
        new_user_vec    = np.mean(self.U[top_similar_idx], axis=0)

        scores = np.clip(new_user_vec @ self.V.T, 0.5, 5.0)
        top_indices = np.argsort(scores)[::-1][:top_n]
        return [(trek_ids[ji], trek_names[trek_ids[ji]], round(scores[ji], 3))
                for ji in top_indices]


# ════════════════════════════════════════════════════════════════════════════
#  STEP 6 — TRAIN THE MODEL
# ════════════════════════════════════════════════════════════════════════════
print("\n" + "═"*65)
print("  STEP 6 : Training the ALS Model")
print("═"*65)

model = ALSRecommender(
    n_factors    = 20,    # 20 hidden features — good balance for 188 treks
    n_iterations = 20,    # 20 ALS rounds — converges well
    lambda_reg   = 0.1,   # regularisation — prevents overfitting
    verbose      = True
)
model.fit(train_matrix)
```

```python
print("\n" + "═"*65)
print("  STEP 7 : Evaluating the Model")
print("═"*65)

test_preds = []
test_actuals = []
for _, row in test_df.iterrows():
    ui = user_idx.get(row['user_id'])
    ti = trek_idx.get(row['trek_id'])
    if ui is not None and ti is not None:
        test_preds.append(model.predict(ui, ti))
        test_actuals.append(row['rating'])

test_rmse = np.sqrt(mean_squared_error(test_actuals, test_preds))
test_mae  = np.mean(np.abs(np.array(test_actuals) - np.array(test_preds)))

print(f"  ✓ Test RMSE : {test_rmse:.4f}  (avg prediction error in stars)")
print(f"  ✓ Test MAE  : {test_mae:.4f}  (avg absolute error in stars)")
print(f"\n  Interpretation:")
print(f"    On a 0–5 star scale, the model's predictions are on average")
print(f"    ±{test_mae:.2f} stars away from the actual ratings.")
if test_rmse < 0.5:
    print(f"    ✓ Excellent accuracy (RMSE < 0.5)")
elif test_rmse < 0.8:
    print(f"    ✓ Good accuracy (RMSE < 0.8)")
else:
    print(f"    △ Acceptable accuracy — consider more training iterations")


# ════════════════════════════════════════════════════════════════════════════
#  STEP 8 — RECOMMENDATIONS FOR EXISTING USERS
# ════════════════════════════════════════════════════════════════════════════
print("\n" + "═"*65)
print("  STEP 8 : Sample Recommendations for Existing Users")
print("═"*65)

sample_users = ['user_0000', 'user_0001', 'user_0007', 'user_0009']
sample_profiles = {u['user_id']: u['profile'] for u in users}

for uid in sample_users:
    profile = sample_profiles[uid]
    recs    = model.recommend(uid, top_n=5)
    print(f"\n  User: {uid}  |  Profile: {profile}")
    print(f"  {'#':<3}  {'Trek Name':<40}  {'Predicted Rating':>16}")
    print(f"  {'─'*3}  {'─'*40}  {'─'*16}")
    for rank, (tid, name, score) in enumerate(recs, 1):
        stars = '★' * int(round(score)) + '☆' * (5 - int(round(score)))
        print(f"  {rank:<3}  {name:<40}  {score:>6.3f}  {stars}")


# ════════════════════════════════════════════════════════════════════════════
#  STEP 9 — RECOMMENDATIONS FOR A NEW USER (Cold Start)
# ════════════════════════════════════════════════════════════════════════════
#
#  Cold Start Problem: A new user has no history.
#  Solution: Ask them their preferences and find similar existing users,
#  then use the average of their latent vectors to represent the new user.
#
print("\n" + "═"*65)
print("  STEP 9 : Recommendations for New Users (Cold Start)")
print("═"*65)

new_user_scenarios = [
    {
        'label'       : 'Beginner trekker, low budget',
        'difficulty'  : 2,
        'budget_max'  : 400,
        'duration_max': 7,
        'fitness'     : 1,
    },
    {
        'label'       : 'Experienced trekker, high altitude enthusiast',
        'difficulty'  : 5,
        'budget_max'  : 2500,
        'duration_max': 18,
        'fitness'     : 4,
    },
    {
        'label'       : 'Moderate trekker, mid budget, 10 days',
        'difficulty'  : 3,
        'budget_max'  : 800,
        'duration_max': 10,
        'fitness'     : 3,
    },
]

for scenario in new_user_scenarios:
    label = scenario.pop('label')
    recs  = model.recommend_for_new_user(scenario, top_n=5)
    print(f"\n  New User : {label}")
    print(f"  Preferences : difficulty={scenario['difficulty']}, "
          f"budget=${scenario['budget_max']}, days≤{scenario['duration_max']}, "
          f"fitness={scenario['fitness']}")
    print(f"  {'#':<3}  {'Trek Name':<40}  {'Score':>6}")
    print(f"  {'─'*3}  {'─'*40}  {'─'*6}")
    for rank, (tid, name, score) in enumerate(recs, 1):
        print(f"  {rank:<3}  {name:<40}  {score:>6.3f}")
    scenario['label'] = label  # restore for clarity


# ════════════════════════════════════════════════════════════════════════════
#  STEP 10 — VISUALISATIONS
# ════════════════════════════════════════════════════════════════════════════
print("\n" + "═"*65)
print("  STEP 10 : Saving Training Plots")
print("═"*65)

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('ALS Matrix Factorization — Trek Recommendation Model',
             fontsize=15, fontweight='bold', color='#1F3864', y=0.98)

# ── Plot 1: Training Loss (RMSE) ──────────────────────────────────────────
ax1 = axes[0, 0]
ax1.plot(range(1, len(model.train_losses)+1), model.train_losses,
         color='#2E75B6', linewidth=2.5, marker='o', markersize=5)
ax1.axhline(y=test_rmse, color='#C00000', linestyle='--',
            linewidth=2, label=f'Test RMSE = {test_rmse:.4f}')
ax1.set_title('Training RMSE per Iteration', fontweight='bold')
ax1.set_xlabel('ALS Iteration'); ax1.set_ylabel('RMSE (lower = better)')
ax1.legend(); ax1.grid(True, alpha=0.3)
ax1.set_facecolor('#F8FBFF')

# ── Plot 2: Predicted vs Actual ratings ───────────────────────────────────
ax2 = axes[0, 1]
sample_n = min(500, len(test_actuals))
idxs = np.random.choice(len(test_actuals), sample_n, replace=False)
ax2.scatter([test_actuals[i] for i in idxs],
            [test_preds[i]   for i in idxs],
            alpha=0.4, color='#375623', s=15)
mn, mx = min(test_actuals), max(test_actuals)
ax2.plot([mn, mx], [mn, mx], 'r--', linewidth=1.5, label='Perfect prediction')
ax2.set_title('Predicted vs Actual Ratings (Test Set)', fontweight='bold')
ax2.set_xlabel('Actual Rating'); ax2.set_ylabel('Predicted Rating')
ax2.legend(); ax2.grid(True, alpha=0.3)
ax2.set_facecolor('#F8FBFF')

# ── Plot 3: Distribution of Interaction Ratings ───────────────────────────
ax3 = axes[1, 0]
ax3.hist(interaction_df['rating'], bins=30, color='#4472C4',
         edgecolor='white', alpha=0.85)
ax3.set_title('Distribution of User-Trek Ratings', fontweight='bold')
ax3.set_xlabel('Rating (0–5)'); ax3.set_ylabel('Frequency')
ax3.grid(True, alpha=0.3, axis='y')
ax3.set_facecolor('#F8FBFF')

# ── Plot 4: Top 10 Most Recommended Treks ────────────────────────────────
ax4 = axes[1, 1]
all_recs = {}
for u in users[:100]:   # sample 100 users
    for tid, name, score in model.recommend(u['user_id'], top_n=5):
        all_recs[name] = all_recs.get(name, 0) + 1
top10 = sorted(all_recs.items(), key=lambda x: x[1], reverse=True)[:10]
names10 = [x[0][:25] + '…' if len(x[0]) > 25 else x[0] for x in top10]
counts10 = [x[1] for x in top10]
bars = ax4.barh(names10[::-1], counts10[::-1], color='#1F3864', alpha=0.85)
ax4.set_title('Top 10 Most Recommended Treks\n(across 100 sampled users)',
              fontweight='bold')
ax4.set_xlabel('Times Recommended')
ax4.grid(True, alpha=0.3, axis='x')
ax4.set_facecolor('#F8FBFF')

plt.tight_layout()
plt.savefig('als_training_results.png', dpi=150, bbox_inches='tight')
print("  ✓ Saved: als_training_results.png")


# ════════════════════════════════════════════════════════════════════════════
#  STEP 11 — HYPERPARAMETER GUIDE
# ════════════════════════════════════════════════════════════════════════════
print("\n" + "═"*65)
print("  STEP 11 : Hyperparameter Tuning Guide")
print("═"*65)
print("""
  The 3 key hyperparameters you can tune:

  1. n_factors (currently = 20)
     ─────────────────────────────────────────────────
     What it is : number of hidden/latent dimensions
     Too LOW    : model is too simple, misses patterns
     Too HIGH   : model overfits (memorises training data)
     Suggested  : try 10, 20, 30, 50
     Rule       : start with √(n_treks) ≈ 14 for 188 treks

  2. n_iterations (currently = 20)
     ─────────────────────────────────────────────────
     What it is : how many ALS training cycles
     Too LOW    : model hasn't converged (high RMSE)
     Too HIGH   : wastes time, no improvement after ~20
     Suggested  : 15–30; watch training RMSE plot flatten

  3. lambda_reg (currently = 0.1)
     ─────────────────────────────────────────────────
     What it is : regularisation penalty
     Too LOW    : overfits (great train RMSE, bad test RMSE)
     Too HIGH   : underfits (both train & test RMSE are high)
     Suggested  : try 0.01, 0.05, 0.1, 0.5, 1.0

  To tune: run the model with different values and pick the
  combination that gives the lowest TEST RMSE.
""")

print("═"*65)
print(f"  ✓ MODEL TRAINING COMPLETE")
print(f"  ✓ Test RMSE : {test_rmse:.4f}   Test MAE : {test_mae:.4f}")
print(f"  ✓ Plot saved: als_training_results.png")
print("═"*65 + "\n")
```
